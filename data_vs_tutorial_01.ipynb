```python
x = [1,2,3]
y = [1,4,9]
c = [10,3,0]

plt.plot(x,y)
plt.plot(x,c)
plt.title("test plot")

plt.xlabel("x")
plt.ylabel("y and c")
plt.legend(["This is y","This is c"])

plt.show()
```

```python
data = pd.read_csv("countries.csv")
```

```python
us = data[data.country == "Afghanistan"]
```

```python
china = data[data.country == "China"]

plt.plot(us.year, us.population/10**6)
plt.plot(china.year, china.population/10**6)

plt.legend(["Afganistan","China"])
plt.xlabel("year")
plt.ylabel("population")

plt.show()
```
