```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1, 2, 3, 4, 5, 7]
y = [1, 4, 7, 8, 9, 11]

# Resize your graph (dpi specifies pixels per inch, when saving probably should use 300 if possible)
# plt.figure(figsize=(5,3), dpi=100)

# add a title (specify font parameter with fontdic)
plt.title("matplotlib tutorial", 
          fontdict={"fontname":"Comic Sans MS", "fontsize":23}
         )
plt.xlabel("X Label", fontdict = {"fontname":"Comic Sans MS"})
plt.ylabel("Y Label", fontdict = {"fontname":"Comic Sans MS"})

plt.xticks([i for i in range(0,10,1)])
plt.yticks([i for i in range(0,15,2)])

# select interval we want to plot points at
x2 = np.arange(0, 4.5, 0.5)

# User shorthand notation
# syntax -> fmt = '[color][marker][line]'

# Line number 01
plt.plot(x, y, 'r.--', label="L01")
# plt.plot(x,y, label = "L01", color="orange", marker = "+", markersize = 10,markeredgecolor = "blue",linestyle = "--")

# Line number 02 with 2 parts
plt.plot(x2[:5], x2[:5]**2, "b*-", label="L02_P01")
plt.plot(x2[4:], x2[4:]**2, "g*-", label="L02_P02") 

plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig("mygraph.png", dpi=300)

plt.show()
```
