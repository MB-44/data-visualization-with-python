```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
labels = ["A","B","C"]
values = [1,4,7]

plt.title("Bar Graph", fontdict={"fontname":"Comic Sans MS", "fontsize":23})

# implement the tick for y axis based on the values
plt.yticks([i for i in range(0,max(values)+1,1)])

bars = plt.bar(labels,values)

bars[0].set_hatch("/")
bars[1].set_hatch("+")
bars[2].set_hatch("-")

plt.show()
```
